#### Import all the libraries that are necessary.

```python
import numpy as np
import pandas as pd
```

#### Open the ` Employee.csv` file and name your dataset `employee`.

```python
employee = pd.read_csv('./data/employee.csv')
employee.columns = [i.lower().replace(' ', '_') for i in employee.columns]
employee.head(10)
```

#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
mean1 = employee.groupby(['department']).agg({'salary':'mean'}).reset_index()
mean1
```

```python
mean2 = employee.pivot_table(index=['department'], values = ['salary']).reset_index()
mean2
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
multi = employee.pivot_table(index=['department'], columns = ['title'], values= ['salary'],aggfunc = ['mean']).fillna('')
multi
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
xp = employee.pivot_table(index=['title'], values = ['salary'], columns=['years']).fillna('')
xp
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
employee.head(10)
```

```python
mean4  = employee.pivot_table(index=['department',], values = ['salary'], columns=['name','title', ], margins=True, margins_name='Mean').fillna('').reset_index() #not exactly what was in the question, but at least we got the information of how many employees are in each department, reading the table from left to right-
mean4
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
describe = employee.pivot_table(index= ['department', 'title'], values = ['salary'], aggfunc = ['median','min','max','std']).reset_index()
describe.head(15)
```

#### Based on your comments, fill the missing values with an appropriate value.

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
total  = employee.pivot_table(index=['department',], values = ['salary'], columns=['name'], margins=True, margins_name='total', aggfunc='sum').fillna('').reset_index() 
total
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
total  = employee.pivot_table(index=['department'], values = ['salary','years'], columns=['title'], margins=True, margins_name='Salary-mean/Years-Total', aggfunc={'years':np.sum,'salary':np.mean}).fillna('').round(2).reset_index()
total
```
